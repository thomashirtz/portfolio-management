```python
from binance import Client
```

```python
client = Client()
tickers = client.get_all_tickers()
tickers
```

```python
symbols = [t['symbol'] for t in tickers if 'BNB' in t['symbol']]
symbols
```

```python
symbols = [t['symbol'] for t in tickers if t['symbol'].endswith('USD')]
symbols
```

```python
for ticker in tickers:
    if ticker['symbol'] in symbols:
        print(ticker)
```

```python
import pandas as pd

symbol = 'BNBUSDT'#, 'ETHEUR']
interval = '1M'
start = "2010-01-01"
end = "2021-07-01"

client = Client()
klines = client.get_historical_klines(symbol, interval, start_str=start, end_str=end)

columns = [
    'open_time', 'open', 'high', 'low', 'close', 'volume',
    'close_time', 'quote_asset_volume', 'number_of_trades',
    'taker_buy_base_asset_volume', 'taker_buy_quote_asset_volume',
    'ignore'
]
dataframe = pd.DataFrame(klines, columns=columns)

import datetime
def get_str_time(unix_time):
    return datetime.datetime.fromtimestamp(int(unix_time)).strftime('%Y_%m_%d')

get_str_time(dataframe['open_time'][0]/1000) 
```
