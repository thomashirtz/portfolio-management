```python
def get_date_start(symbol):
    from binance import Client
    import datetime
    interval = '1M'
    start = "2010-01-01"
    end = "2021-07-01"
    index_open_time = 0
    client = Client()
    klines = client.get_historical_klines(symbol, interval, start_str=start, end_str=end)
    unix_time = klines[0][index_open_time] / 1000
    return datetime.datetime.fromtimestamp(unix_time) # .strftime('%Y_%m_%d')

get_date_start('BNBUSDT')
```

```python
symbols = ['EURUSDT', 'ETHBTC', 'USDCUSDT', 'BNBUSDT', 'LTCUSDT', 'XRPUSDT']
for symbol in symbols:
    date = get_date_start(symbol)
    print(f'{symbol} => {date}')
```
