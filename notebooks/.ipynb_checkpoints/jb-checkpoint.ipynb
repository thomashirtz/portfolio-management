````python
def preprocessed(
        dataset: xr.Dataset,
        log_columns: List[str] = None,
        keep_columns: List[str] = None,
        index_column: bool = True,
        sampling_rate: Optional[int] = None,
        **_
) -> xr.Dataset:

    if sampling_rate:
        dataset = dataset[{MEASUREMENT_INDEX: slice(None, None, sampling_rate)}]

    log_columns_ = []
    if log_columns:
        def insert_after_substring(string, string_to_insert, substring):
            index = string.find(substring)
            return string[:index + len(substring)] + string_to_insert + string[len(substring) + index:]

        log_columns_ = [insert_after_substring(c, 'log_', '#') for c in log_columns]
        logged = np.log(dataset[[MEASUREMENT]].sel({CHARACTERISTIC_NAME: log_columns}))
        logged = logged.assign_coords({CHARACTERISTIC_NAME: log_columns_})
        dataset = xr.concat([dataset, logged], dim=CHARACTERISTIC_NAME, data_vars="minimal")

    index_column_ = [] # todo quite complicated for what it is
    if index_column:
        index_column_ = [MEASUREMENT_INDEX]
        index_dataarray = dataset[[MEASUREMENT]].isel({CHARACTERISTIC_NAME: 0})
        index_dataarray[MEASUREMENT].values = np.tile(
            np.arange(dataset[MEASUREMENT].shape[2]),
            (dataset[MEASUREMENT].shape[0], 1)
        ).astype(np.float32)
        index_dataarray = index_dataarray.assign_coords({CHARACTERISTIC_NAME: index_column_})
        index_dataset = xr.Dataset(index_dataarray)
        dataset = xr.concat([dataset, index_dataset], dim=CHARACTERISTIC_NAME, data_vars="minimal")

    keep_columns = [] if keep_columns is None else keep_columns
    keep_columns_ = keep_columns + log_columns_ + index_column_
    columns_characteristic = list(dataset[CHARACTERISTIC_NAME].values)
    columns_drop_sel = list(filter(lambda x: x not in keep_columns_, columns_characteristic))

    return dataset.drop_sel({CHARACTERISTIC_NAME: columns_drop_sel})
````
